## Variáveis

**O que são variáveis?**

As variáveis em Python servem para facilitar a manipulação de valores dentro de um programa.
Cada variável é responsável por armazenar um valor, os quais podem ser mudados e atualizados durante o programa.
Por existirem diferentes tipos de valores que podem ser atribuídos a uma variável (como números, texto, verdadeiro/falso), existem diferentes tipos de variáveis dentro do Python.

* **Inteiros: `int()`** --> Números inteiros, ou seja aqueles que não possuem vírgula.

```python
print(int(1))
print(int(1.8423)) #não armazena decimais
```

* **Decimal - `float()`** --> Números decimais, ou seja aqueles que são racionais e portanto possuem vírgula.

```python
print(float(1))
print(float(1.8423))
```

* **String - `str()`** --> Usado para representar em geral coisas escritas como palavras, frases e textos. É representado sempre pelas aspas ('  ' ou "  ").

```python
print("hello world")
print('hello world')
```

* **Boleano - `bool`** --> Representa um tipo de dado que só pode ser classificado como verdadeiro (True) ou falso (False).

* **Listas - `list()`** --> Variável que representa uma lista de elementos variados, ou seja, uma lista que pode conter todos os elementos descritos acima. As listas são delimitadas por [  ] e os elementos dentro dela são separados por vírgulas (,).

* **Tuplas - `tuple()`** --> São listas imutáveis.
As tuplas são delimitadas por (  ) e igual as listas tem seus elementos separados por vírgulas (,).

* **Dicionários - `dict()`** --> Dicionário é uma variável capaz de associar dois elementos, com o sistema de chave e valor.

Para usarmos uma variável, nós declaramos ela de um jeito bem simples:

```python
a = 2
b = 'Hello World!'
```

Agora toda vez que usarmos a variável a ou b dentro de um código, será a mesma coisa que usar o valor numérico 2 ou o texto 'Hello World'':

```python
print(a)
print(b)
```

É possível também mudar os valores de variáveis e executar operações:

```python
a = 4 #agora a variável passa a ter o valor 4
b = 6

c = b + a

print(c)
```

## Condicionais

**O que são condicionais em Python?**

Na aula de hoje veremos como desenvolver programas que, dada uma condição, executam, ou não, um determinado bloco de código. Suponha que queremos desenvolver uma função que recebe dois números, x e y como argumento e devolve 1 se x for maior do que y e 0, caso contrário. A não ser que exista uma fórmula fechada (existe, para alguns casos), precisamos de alguma forma de executar um trecho de código **somente se** x for maior que y e outro trecho de código **somente se** essa condição não for verdadeira.

Para isso vamos precisar do operador if. Vamos começar com um exemplo curto para entender como ele funciona.

```python
lampada_acesa = True

if lampada_acesa:
    print("Está muito claro aqui!")
    lampada_acesa = False

if lampada_acesa:
    print("Continua muito claro aqui!")
```

O operador if funciona da seguinte maneira:

* No caso de if True: o bloco do if é executado;
* No caso de if False: o bloco do if não é executado.

No caso acima, o primeiro bloco do if é executado, já que o valor de lampada_acesa é True. Entretanto, no fim desse bloco, seu valor muda para false, e por isso o segundo bloco não é executado

O operador else é complementar ao if, representando o caso contrário ao mesmo. Ele captura todos os casos para os quais a condição do if é falsa. Um exemplo de seu uso está abaixo:


```python
numero = int(input('Digite um número: '))

if numero == 86:
    print('Resposta para a vida, o universo e tudo mais')
else:
    print('Um número qualquer')
```

No caso, se o input do usuário for exatamente 86, o primeiro bloco, do if, será executado. Caso contrário, o segundo bloco, do else, entrará em cena


Por último há um operador chamado elif. Esse operador é utilizado para realizar a verificação de outra expressão caso a primeira validação seja falsa

```python
idade = int(input('Digite a sua idade: '))

if idade < 12:
  print('Você é criança')
elif idade < 20:
  print('Você é adolescente')
elif idade < 60:
  print('Você é adulto')
else:
  print('Você é idoso')
```

Assim, quando a idade colocada for menor que 12, o programa irá dizer que a pessoa é criança. Caso essa expressão seja falsa (a idade é superior a 12) o programa continuará lendo o código até encontrar alguma expressão que seja verdadeira. Caso nenhuma expressão seja verdadeira o programa irá printar o que está no else, que no caso é "Você é idoso".

## Listas

**O que é uma lista em Python?**

Como já foi definido na aula de variáveis, as listas em Python são um tipo de variável que representa uma lista de elementos variados, ou seja, uma lista que pode conter todos os elementos descritos acima.


**As listas são delimitadas por [  ] e os elementos dentro dela são separados por vírgulas (,).**


Assim como as strings, cada elemento de uma lista possui um índice e é por meio dele que podemos acessar aquele elemento, nos permitindo até fazer alterações sobre o mesmo.

**A contagem dos índices de uma lista sempre se inicia em 0.**

**Acessando elementos de uma Lista**

**lista_exemplo = ['a' , 'b', 'c', 'd']**

* lista com 4 elementos
* índices: 0, 1, 2, 3


```python
lista_exemplo =  ['a' , 'b', 'c', 'd']
print(lista_exemplo[1])
```

**Como manipular listas?**

As listas são um tipo de variável que permite diversas formas de manipulação e portanto pode-se fazer muitas coisas com ela.

**Quantidade de elementos de uma lista - len( lista )**


Usar **.len(lista)** para encontrar quantos elementos tem dentro dela. O retorno dessa função é sempre um número inteiro.

```python
informacoes_pessoais = ['Sergio', 29, 11900003333]
a = len(informacoes_pessoais)
print(f'A lista tem {a} elementos')
```

**Achar qual o índice de x elemento dentro de uma lista**

Usar **.index(elemento)** - para achar qual o índice da lista que aquele elemento está

Se o elemento aparecer mais de uma vez na lista, essa função deve retornar o primeiro índice em que ele aparece

```python
informacoes_pessoais = ['Sergio', 29, 11900003333]
b = informacoes_pessoais.index('Sergio')
print(f'A informação do nome está em {b}')
```

```python
informacoes_pessoais.append('endereço')
print(informacoes_pessoais)
```

**Removendo um elemento de uma lista**

**Removendo elemento qualquer - `.pop( indice )`**

* se usado sem especificar qual indice deve ser removido, ele irá apagar o último elemento

```python
nomes = ['Joao', 'Maria', 'Gabriela', 'Fernando']
nomes.pop(0)
print(nomes)
```

**Removendo elemento escolhido - `.remove(elemento)`**

* deve especificar qual o elemento que deve ser removido e essa função irá apagar da lista a primeira ocorrência do mesmo dentro dela

```python
nomes.remove('Gabriela')
print(nomes)
```

**Descobrir se certo elemento está dentro de uma lista ou não**

elemento `in lista`

## While ("enquanto")


**O que é e como funciona o While?**

**O while funciona de maneira similar ao if.** Entretanto, ao chegar no final de um bloco de código desse tipo, o programa **reavalia se a condição inicialmente proposta ainda está satisfeita.**

* Um exemplo de um programa deste tipo está exposto abaixo:

```python
tem_duvidas = True
while tem_duvidas:
    resposta_do_aluno = input('Você está com dúvidas? ')
    if resposta_do_aluno == 'não':
        print('Até a próxima')
        tem_duvidas = False
    else:
        print('Pratique mais')
```

Neste exemplo, enquanto tem_duvidas for verdadeiro, ou seja, o aluno não responder com a palavra 'não', o programa continua perguntando 'você está com dúvidas?'.
Assim que a resposta for negativa, a condição, de tem_duvidas == True, deixa de ser satisfeita e o bloco deixa de ser executado.

**O while pode também ser usado como contador. Um exemplo disso é colocado abaixo:**



```python
contador = 0
while contador < 4:
    print(contador)
    contador += 1
```

Nele, a variável 'contador' vai sendo acrescida em uma unidade até chegar ao valor 3, quando o bloco condicional deixa de ser executado.

## For/Range

O for pode ser usado em situações semelhantes ao while. Entretanto, não é necessário criar uma variável, cuja veracidade é avaliada  cada iteração do bloco de código, anteriormente, como no while. No caso desse comando, para realizar tarefas repetitivas, com uma quantidade pré-determinada de iterações, é utilizado o comando range junto com a chamada da função for, junto com um contador qualquer, como no bloco de código abaixo:


```python
for i in range(0, 3):
  print(i)
```

Na maioria das vezes, o uso do for é mais eficiente que o uso do while, já que é necessário menos linhas de código para realizar a mesma tarefa. Os argumentos a ela fornecido são, nessa ordem, o número de início, que é contabilizado, e o de final, que não é levado em consideração - a contagem para um passo antes. Além disso, pode ser fornecido um terceiro argumento, o passo. Ele determina o intervalo entre os valores do contador em cada iteração do bloco de código. Caso ele não seja fornecido, o passo será 1.


Outro uso importante da função for é percorrer itens em uma lista. Isso é evidenciado no exemplo abaixo:


```python
lista1 = ['Joao', 'Alberto', 'Pedro']
for item in lista1:
    print(item)
```

No código acima, cada nome da lista fornecida será impresso no terminal. Esse uso do for é muito útil e eficiente para manipular e extrair dados fornecidos dessa maneira.

## Funções

**O que são Funções em Python?**

Uma função em Python é um bloco de **código definido (ou seja, que tem um nome) que realiza alguma tarefa e pode ser executado várias vezes ao longo de todo o código** se ele for chamado.
As funções servem para facilitar a programação, ajudando a evitar a repetição de blocos de código, a existência de códigos grandes e confusos.
As funções precisam ser definidas e atuam sobre qualquer tipo de variável, desde que os seus comandos respeitem os limites dessa variável. **Todas as funções devem retornar algo**, que por sua vez também pode ser qualquer tipo de variável.
Ou seja, elas **funcionam como um cozinheiro executando uma receita de bolo**. Você fornece para ele os ingredientes e no final ele te retorna o bolo pronto.

**Como montar uma função?**

Para declarar uma função precisamos seguir alguns passos e passar algumas informações ao Python para que ela seja definida corretamente e possa ser usada ao longo de todo o código abaixo dela.


```python
def soma_numeros(n1, n2):
  soma = n1 + n2
  return soma

a = soma_numeros(3, 4)
print(a)
```

* **Declarar a função - linha 1 do exemplo acima**

(def  nome_da_função( parâmetros ) ):     - nunca esqueça de colocar o ":"


Para declarar uma função precisamos usar o def (de definir), declarar após ele o nome dela e, caso ela receba algum parâmetro, o nome dos seus parâmetros.


Os parâmetros são informações sobre as quais o código da função pode usar para produzir o seu resultado e os nomes deles são como eles serão referidos dentro do código da função.


* **Corpo da função - linha 2 do exemplo acima**

O corpo da função é basicamente o código que ela deve executar, ou seja, o que deve acontecer dentro dela.




* **Retorno da função (return) - linha 3 do exemplo acima**

Dentro de toda função devemos declarar o que ela deve retornar uma vez que foi executada, ou seja, ele deve apontar o (ou os) produto(s) daquela função.


**Exemplo de funções**

* Achar primeira letra

```python
def primeira_letra(palavra):
  pl = palavra[0]
  return pl

teste = primeira_letra('televisão')
print(teste)
```

* Transforar altura de cm em pés (unidade de medida)


```python
def altura_pes(altura_cm):
  em_pes = altura_cm * 0.0328084
  return em_pes

altura_pedro = 183
print(altura_pes(altura_pedro))
```
